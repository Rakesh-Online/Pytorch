<h1 style="color:Tomato;">What is Pytorch ?</h1>
<h2 style="background-color:powderblue;">PyTorch is an open source library developed mainly by Facebook's artificial intelligence research group as a Python version of Torch.</h2>
<h2>PyTorch was first released to the public in January 2017. It uses the power of GPUs to speed up the computation of tensors, which accelerates the training times of complex models.</h2>
<h2>The library has a C++ backend, combined with the deep learning framework of Torch, which allows much faster computations than native Python libraries with many deep learning features. </h2>
<h2>PyTorch also supports distributed training which enables researchers as well as practitioners to parallelize their computations. Distributed training makes it possible to use multiple GPUs to process larger batches of input data. This, in turn, reduces the computation time</h2>


<h1 style="color:Tomato;">What is Tensors ? </h1>
<h2 style="background-color:powderblue;">A tensor is a generic n-dimensional array, very similar to numpy arrays. The difference being that PyTorch tensors can run on the GPU (while NumPy tensors cannot),  which helps to accelerate numerical computations.</h2> 
<h2> Similar to NumPy arrays or any other matrix-like structure, PyTorch tensors can have as many dimensions as desired.</h2>

<h2>For tensors, dimensions are also known as ranks. </h2>
